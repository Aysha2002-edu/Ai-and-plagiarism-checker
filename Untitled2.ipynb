```python
# =========================================================
# SECTION 3: UTILITY FUNCTIONS
# =========================================================

def find_file_recursively(root_dir, filename):
    root_dir = Path(root_dir)
    matches = list(root_dir.rglob(filename))
    if not matches:
        raise FileNotFoundError(f"Could not find {filename} inside {root_dir}")
    return matches[0]

def load_jsonl(path):
    rows = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                rows.append(json.loads(line))
    return pd.DataFrame(rows)

def infer_text_column(df):
    priority = ["text", "document", "content", "article", "essay", "body"]
    for c in priority:
        if c in df.columns:
            return c

    object_cols = [c for c in df.columns if df[c].dtype == "object"]
    if not object_cols:
        raise ValueError("No text-like column found.")

    scores = {}
    for c in object_cols:
        try:
            lengths = df[c].dropna().astype(str).str.len()
            scores[c] = lengths.mean()
        except Exception:
            scores[c] = -1
    return max(scores, key=scores.get)

def infer_label_column(df, text_col):
    priority = ["label", "class", "target", "gold", "generated", "source"]
    for c in priority:
        if c in df.columns and c != text_col:
            return c

    candidate_cols = [c for c in df.columns if c != text_col]
    if not candidate_cols:
        raise ValueError("No label-like column found.")

    uniq_counts = {}
    for c in candidate_cols:
        try:
            uniq_counts[c] = df[c].nunique(dropna=True)
        except Exception:
            uniq_counts[c] = 999999

    return sorted(uniq_counts.items(), key=lambda x: x[1])[0][0]

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = re.sub(r"<[^>]+>", " ", text)
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def simple_tokenize(text):
    if not isinstance(text, str):
        return []
    return re.findall(r"\b\w+\b", text.lower())

def token_count(text):
    return len(simple_tokenize(text))

def sentence_split(text):
    if not text or not isinstance(text, str):
        return []
    doc = nlp(text)
    return [sent.text.strip() for sent in doc.sents if sent.text.strip()]

def ensure_binary_label(label):
    """
    0 -> Human
    1 -> AI-generated / AI-involved
    """
    if pd.isna(label):
        return np.nan

    if isinstance(label, (int, float, np.integer, np.floating)):
        label_int = int(label)
        if label_int in [0, 1]:
            return label_int
        return 0 if label_int == 0 else 1

    s = str(label).strip().lower()

    human_strings = {
        "human", "fully_human", "human_written", "human-written",
        "no_ai", "non_ai", "original_human", "pure_human"
    }
    if s in human_strings:
        return 0

    if "human" in s and all(x not in s for x in ["ai", "gpt", "generated", "rewrite", "paraphrase", "summary", "translate"]):
        return 0

    return 1

def stratified_split(df, train_frac=0.8, val_frac=0.1, test_frac=0.1, seed=42):
    assert abs(train_frac + val_frac + test_frac - 1.0) < 1e-8
    rng = np.random.RandomState(seed)

    train_parts, val_parts, test_parts = [], [], []

    for _, group in df.groupby("label_bin"):
        idx = group.index.to_numpy()
        rng.shuffle(idx)

        n = len(idx)
        n_train = int(round(n * train_frac))
        n_val = int(round(n * val_frac))
        n_test = n - n_train - n_val

        if n >= 3:
            if n_train == 0:
                n_train = 1
            if n_val == 0:
                n_val = 1
            if n_test == 0:
                n_test = 1
            while n_train + n_val + n_test > n:
                if n_train > 1:
                    n_train -= 1
                elif n_val > 1:
                    n_val -= 1
                else:
                    n_test -= 1

        train_idx = idx[:n_train]
        val_idx = idx[n_train:n_train+n_val]
        test_idx = idx[n_train+n_val:]

        train_parts.append(df.loc[train_idx])
        val_parts.append(df.loc[val_idx])
        test_parts.append(df.loc[test_idx])

    train_df = pd.concat(train_parts).sample(frac=1, random_state=seed).reset_index(drop=True)
    val_df = pd.concat(val_parts).sample(frac=1, random_state=seed).reset_index(drop=True)
    test_df = pd.concat(test_parts).sample(frac=1, random_state=seed).reset_index(drop=True)

    return train_df, val_df, test_df

def print_class_stats(df, name="Dataset"):
    print(f"\n{name} size: {len(df):,}")
    counts = df["label_bin"].value_counts().sort_index()
    for cls in [0, 1]:
        print(f"  Class {cls} ({'Human' if cls == 0 else 'AI'}): {counts.get(cls, 0):,}")

def describe_lengths(df, text_col="text_clean"):
    lengths = df[text_col].apply(token_count)
    print("\nText length statistics (tokens):")
    print(lengths.describe(percentiles=[0.25, 0.5, 0.75, 0.9, 0.95]).to_string())
```

```python
# =========================================================
# SECTION 8: METRICS & PLOTTING UTILITIES
# =========================================================

def confusion_matrix_binary(y_true, y_pred):
    y_true = np.array(y_true).astype(int)
    y_pred = np.array(y_pred).astype(int)
    tp = int(((y_true == 1) & (y_pred == 1)).sum())
    tn = int(((y_true == 0) & (y_pred == 0)).sum())
    fp = int(((y_true == 0) & (y_pred == 1)).sum())
    fn = int(((y_true == 1) & (y_pred == 0)).sum())
    return np.array([[tn, fp], [fn, tp]])

def roc_curve_auc(y_true, y_score):
    y_true = np.array(y_true).astype(int)
    y_score = np.array(y_score).astype(float)

    order = np.argsort(-y_score)
    y_true = y_true[order]

    P = (y_true == 1).sum()
    N = (y_true == 0).sum()

    if P == 0 or N == 0:
        return np.array([0, 1]), np.array([0, 1]), 0.5

    tpr = [0.0]
    fpr = [0.0]
    tp = 0
    fp = 0

    for i in range(len(y_true)):
        if y_true[i] == 1:
            tp += 1
        else:
            fp += 1
        tpr.append(tp / P)
        fpr.append(fp / N)

    tpr.append(1.0)
    fpr.append(1.0)

    auc = np.trapz(np.array(tpr), np.array(fpr))
    return np.array(fpr), np.array(tpr), float(auc)

def classification_metrics(y_true, y_prob, threshold=0.5):
    y_true = np.array(y_true).astype(int)
    y_prob = np.array(y_prob).astype(float)
    y_pred = (y_prob >= threshold).astype(int)

    cm = confusion_matrix_binary(y_true, y_pred)
    tn, fp = cm[0]
    fn, tp = cm[1]

    accuracy = (tp + tn) / max((tp + tn + fp + fn), 1)
    precision = tp / max((tp + fp), 1)
    recall = tp / max((tp + fn), 1)
    f1 = (2 * precision * recall) / max((precision + recall), 1e-12)

    fpr, tpr, auc = roc_curve_auc(y_true, y_prob)

    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "roc_auc": auc,
        "confusion_matrix": cm,
        "fpr": fpr,
        "tpr": tpr,
        "y_pred": y_pred
    }

def plot_confusion_matrix(cm, title="Confusion Matrix"):
    plt.figure(figsize=(4, 4))
    plt.imshow(cm, interpolation='nearest')
    plt.title(title)
    plt.colorbar()
    plt.xticks([0, 1], ["Human", "AI"])
    plt.yticks([0, 1], ["Human", "AI"])
    plt.xlabel("Predicted")
    plt.ylabel("True")

    for i in range(2):
        for j in range(2):
            plt.text(j, i, cm[i, j], ha="center", va="center", color="black", fontsize=12)

    plt.tight_layout()
    plt.show()

def plot_roc_curves(curves_dict, title="ROC Curves"):
    plt.figure(figsize=(7, 5))
    for name, (fpr, tpr, auc) in curves_dict.items():
        plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.4f})")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

def print_metrics_table(name, metrics_dict):
    print(f"\n{name}")
    print("-" * 50)
    print(f"Accuracy : {metrics_dict['accuracy']:.4f}")
    print(f"Precision: {metrics_dict['precision']:.4f}")
    print(f"Recall   : {metrics_dict['recall']:.4f}")
    print(f"F1-score : {metrics_dict['f1']:.4f}")
    print(f"ROC-AUC  : {metrics_dict['roc_auc']:.4f}")
```

```python
# =========================================================
# SECTION 10: FAST STYLOMETRY FEATURE EXTRACTION
# =========================================================

PUNCTUATIONS = [".", ",", ";", ":", "!", "?", "'", '"', "(", ")", "-"]

POS_BUCKETS = {
    "NOUN": "noun",
    "PROPN": "noun",
    "VERB": "verb",
    "AUX": "verb",
    "ADJ": "adj",
    "ADV": "adv",
    "PRON": "pron",
    "DET": "det",
    "ADP": "adp",
    "CCONJ": "conj",
    "SCONJ": "conj",
    "NUM": "num"
}

SPACY_BATCH_SIZE = 512
SPACY_N_PROCESS = 2

def safe_mean(x):
    return float(np.mean(x)) if len(x) > 0 else 0.0

def safe_std(x):
    return float(np.std(x)) if len(x) > 0 else 0.0

def stylometry_features_from_doc(doc, original_text):
    sentences = [sent for sent in doc.sents if sent.text.strip()]
    words = [token.text for token in doc if token.is_alpha]
    words_lower = [w.lower() for w in words]

    sent_word_lens = []
    for sent in sentences:
        sent_words = [token.text for token in sent if token.is_alpha]
        sent_word_lens.append(len(sent_words))

    word_lens = [len(w) for w in words]

    n_chars = max(len(original_text), 1)
    n_words = max(len(words), 1)

    feats = {}
    feats["avg_sentence_len"] = safe_mean(sent_word_lens)
    feats["std_sentence_len"] = safe_std(sent_word_lens)
    feats["avg_word_len"] = safe_mean(word_lens)
    feats["std_word_len"] = safe_std(word_lens)
    feats["type_token_ratio"] = len(set(words_lower)) / n_words
    feats["hapax_ratio"] = sum(v == 1 for v in Counter(words_lower).values()) / n_words

    for p in PUNCTUATIONS:
        key = f"punct_{repr(p)[1:-1]}"
        feats[key] = original_text.count(p) / n_chars

    feats["digit_ratio"] = sum(ch.isdigit() for ch in original_text) / n_chars
    feats["upper_ratio"] = sum(ch.isupper() for ch in original_text) / n_chars
    feats["space_ratio"] = sum(ch.isspace() for ch in original_text) / n_chars

    pos_counts = defaultdict(int)
    pos_total = 0

    for token in doc:
        if token.is_space:
            continue
        bucket = POS_BUCKETS.get(token.pos_, "other")
        pos_counts[bucket] += 1
        pos_total += 1

    pos_names = ["noun", "verb", "adj", "adv", "pron", "det", "adp", "conj", "num", "other"]
    pos_total = max(pos_total, 1)

    for p in pos_names:
        feats[f"pos_{p}"] = pos_counts[p] / pos_total

    return feats

def extract_stylometry_features_spacy(texts, batch_size=512, n_process=2):
    texts = [str(t) for t in texts]
    features = []
    pipe_iter = nlp.pipe(texts, batch_size=batch_size, n_process=n_process)

    for doc, original_text in tqdm(
        zip(pipe_iter, texts),
        total=len(texts),
        desc="Stylometry extraction"
    ):
        features.append(stylometry_features_from_doc(doc, original_text))

    return pd.DataFrame(features)

train_stylo_df = extract_stylometry_features_spacy(
    train_df_fast["text_clean"].tolist(),
    batch_size=SPACY_BATCH_SIZE,
    n_process=SPACY_N_PROCESS
)

val_stylo_df = extract_stylometry_features_spacy(
    val_df_fast["text_clean"].tolist(),
    batch_size=SPACY_BATCH_SIZE,
    n_process=SPACY_N_PROCESS
)

test_stylo_df = extract_stylometry_features_spacy(
    test_df_fast["text_clean"].tolist(),
    batch_size=SPACY_BATCH_SIZE,
    n_process=SPACY_N_PROCESS
)

feature_cols = train_stylo_df.columns.tolist()
print("Number of stylometric features:", len(feature_cols))
```

```python
# =========================================================
# SECTION 11: STYLOMETRY MODEL
# =========================================================

class StandardScalerTorch:
    def __init__(self):
        self.mean_ = None
        self.std_ = None

    def fit(self, X):
        X = np.asarray(X, dtype=np.float32)
        self.mean_ = X.mean(axis=0)
        self.std_ = X.std(axis=0)
        self.std_[self.std_ < 1e-8] = 1.0
        return self

    def transform(self, X):
        X = np.asarray(X, dtype=np.float32)
        return (X - self.mean_) / self.std_

    def fit_transform(self, X):
        return self.fit(X).transform(X)

class NumpyDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(np.asarray(X), dtype=torch.float32)
        self.y = torch.tensor(np.asarray(y), dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

class SimpleMLPBinary(nn.Module):
    def __init__(self, input_dim, hidden_dim=32, dropout=0.2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, x):
        return self.net(x).squeeze(-1)

def train_binary_torch_model(model, train_loader, val_loader, epochs=5, lr=1e-3, pos_weight=None):
    model = model.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    if pos_weight is not None:
        pos_weight = torch.tensor([pos_weight], dtype=torch.float32, device=DEVICE)
        criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
    else:
        criterion = nn.BCEWithLogitsLoss()

    best_state = None
    best_val_f1 = -1

    for epoch in range(epochs):
        model.train()
        train_losses = []

        for xb, yb in train_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())

        model.eval()
        val_probs = []
        val_true = []

        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(DEVICE)
                logits = model(xb)
                probs = torch.sigmoid(logits).cpu().numpy()
                val_probs.extend(probs.tolist())
                val_true.extend(yb.numpy().tolist())

        val_metrics = classification_metrics(val_true, val_probs)
        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {np.mean(train_losses):.4f} | Val F1: {val_metrics['f1']:.4f}")

        if val_metrics["f1"] > best_val_f1:
            best_val_f1 = val_metrics["f1"]
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}

    model.load_state_dict(best_state)
    return model

def predict_binary_torch_model(model, X, batch_size=256):
    model.eval()
    probs = []
    ds = torch.tensor(np.asarray(X), dtype=torch.float32)
    loader = DataLoader(ds, batch_size=batch_size, shuffle=False)

    with torch.no_grad():
        for xb in loader:
            xb = xb.to(DEVICE)
            logits = model(xb)
            batch_probs = torch.sigmoid(logits).cpu().numpy()
            probs.extend(batch_probs.tolist())
    return np.array(probs)

X_train_stylo = train_stylo_df.values.astype(np.float32)
X_val_stylo = val_stylo_df.values.astype(np.float32)
X_test_stylo = test_stylo_df.values.astype(np.float32)

scaler = StandardScalerTorch()
X_train_stylo_sc = scaler.fit_transform(X_train_stylo)
X_val_stylo_sc = scaler.transform(X_val_stylo)
X_test_stylo_sc = scaler.transform(X_test_stylo)

n_pos = (y_train_fast == 1).sum()
n_neg = (y_train_fast == 0).sum()
pos_weight = n_neg / max(n_pos, 1)

train_loader_stylo = DataLoader(NumpyDataset(X_train_stylo_sc, y_train_fast), batch_size=128, shuffle=True)
val_loader_stylo = DataLoader(NumpyDataset(X_val_stylo_sc, y_val_fast), batch_size=256, shuffle=False)

stylometry_model = SimpleMLPBinary(input_dim=X_train_stylo_sc.shape[1], hidden_dim=32, dropout=0.2)
stylometry_model = train_binary_torch_model(
    stylometry_model,
    train_loader_stylo,
    val_loader_stylo,
    epochs=5,
    lr=1e-3,
    pos_weight=pos_weight
)

stylometry_train_score = predict_binary_torch_model(stylometry_model, X_train_stylo_sc)
stylometry_val_score = predict_binary_torch_model(stylometry_model, X_val_stylo_sc)
stylometry_test_score = predict_binary_torch_model(stylometry_model, X_test_stylo_sc)

stylometry_metrics = classification_metrics(y_test_fast, stylometry_test_score)
print_metrics_table("Stylometry Model", stylometry_metrics)
```

```python
# =========================================================
# SECTION 12: FAST SEMANTIC MODEL (RoBERTa)
# =========================================================

MODEL_NAME = "roberta-base"
MAX_LEN = 96
BATCH_SIZE = 32
EPOCHS = 1
LR = 2e-5

class TextClassificationDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=96):
        self.texts = list(texts)
        self.labels = list(labels)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])

        enc = self.tokenizer(
            text,
            truncation=True,
            max_length=self.max_length,
            padding=False,
            return_tensors=None
        )
        enc["labels"] = label
        return enc

semantic_tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
semantic_model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=2
).to(DEVICE)

# freeze embeddings and first 9 transformer layers for speed
for param in semantic_model.roberta.embeddings.parameters():
    param.requires_grad = False

for layer in semantic_model.roberta.encoder.layer[:9]:
    for param in layer.parameters():
        param.requires_grad = False

train_ds_sem = TextClassificationDataset(
    train_df_fast["text_clean"],
    train_df_fast["label_bin"],
    semantic_tokenizer,
    max_length=MAX_LEN
)
val_ds_sem = TextClassificationDataset(
    val_df_fast["text_clean"],
    val_df_fast["label_bin"],
    semantic_tokenizer,
    max_length=MAX_LEN
)
test_ds_sem = TextClassificationDataset(
    test_df_fast["text_clean"],
    test_df_fast["label_bin"],
    semantic_tokenizer,
    max_length=MAX_LEN
)

collator = DataCollatorWithPadding(tokenizer=semantic_tokenizer)

train_loader_sem = DataLoader(train_ds_sem, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collator)
val_loader_sem = DataLoader(val_ds_sem, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collator)
test_loader_sem = DataLoader(test_ds_sem, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collator)

optimizer = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, semantic_model.parameters()),
    lr=LR
)

total_steps = len(train_loader_sem) * EPOCHS
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=max(1, int(0.05 * total_steps)),
    num_training_steps=total_steps
)

class_counts = train_df_fast["label_bin"].value_counts().to_dict()
count_0 = class_counts.get(0, 1)
count_1 = class_counts.get(1, 1)
class_weights = torch.tensor([1.0 / count_0, 1.0 / count_1], dtype=torch.float32, device=DEVICE)
class_weights = class_weights / class_weights.sum() * 2.0

amp_enabled = torch.cuda.is_available()
scaler_amp = torch.cuda.amp.GradScaler(enabled=amp_enabled)

def evaluate_semantic(model, loader):
    model.eval()
    probs = []
    true = []
    losses = []

    with torch.no_grad():
        for batch in loader:
            labels = batch["labels"].to(DEVICE)
            batch_inputs = {k: v.to(DEVICE) for k, v in batch.items() if k != "labels"}

            with torch.cuda.amp.autocast(enabled=amp_enabled):
                outputs = model(**batch_inputs)
                logits = outputs.logits
                loss = F.cross_entropy(logits, labels, weight=class_weights)

            pred_prob = F.softmax(logits, dim=1)[:, 1].detach().cpu().numpy()
            probs.extend(pred_prob.tolist())
            true.extend(labels.detach().cpu().numpy().tolist())
            losses.append(loss.item())

    metrics = classification_metrics(true, probs)
    return np.mean(losses), metrics, np.array(probs)

best_state = None
best_val_f1 = -1

for epoch in range(EPOCHS):
    semantic_model.train()
    train_losses = []

    for batch in tqdm(train_loader_sem, desc=f"Semantic Epoch {epoch+1}/{EPOCHS}"):
        labels = batch["labels"].to(DEVICE)
        batch_inputs = {k: v.to(DEVICE) for k, v in batch.items() if k != "labels"}

        optimizer.zero_grad()

        with torch.cuda.amp.autocast(enabled=amp_enabled):
            outputs = semantic_model(**batch_inputs)
            logits = outputs.logits
            loss = F.cross_entropy(logits, labels, weight=class_weights)

        scaler_amp.scale(loss).backward()
        scaler_amp.step(optimizer)
        scaler_amp.update()
        scheduler.step()

        train_losses.append(loss.item())

    val_loss, val_metrics, _ = evaluate_semantic(semantic_model, val_loader_sem)
    print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {np.mean(train_losses):.4f} | Val F1: {val_metrics['f1']:.4f}")

    if val_metrics["f1"] > best_val_f1:
        best_val_f1 = val_metrics["f1"]
        best_state = {k: v.cpu().clone() for k, v in semantic_model.state_dict().items()}

semantic_model.load_state_dict(best_state)

def predict_semantic_probs(model, loader):
    model.eval()
    probs = []
    true = []

    with torch.no_grad():
        for batch in loader:
            labels = batch["labels"].to(DEVICE)
            batch_inputs = {k: v.to(DEVICE) for k, v in batch.items() if k != "labels"}

            with torch.cuda.amp.autocast(enabled=amp_enabled):
                outputs = model(**batch_inputs)
                logits = outputs.logits

            pred_prob = F.softmax(logits, dim=1)[:, 1].detach().cpu().numpy()
            probs.extend(pred_prob.tolist())
            true.extend(labels.detach().cpu().numpy().tolist())

    return np.array(probs), np.array(true)

semantic_train_loader_eval = DataLoader(train_ds_sem, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collator)

semantic_train_score, _ = predict_semantic_probs(semantic_model, semantic_train_loader_eval)
semantic_val_score, _ = predict_semantic_probs(semantic_model, val_loader_sem)
semantic_test_score, _ = predict_semantic_probs(semantic_model, test_loader_sem)

semantic_metrics = classification_metrics(y_test_fast, semantic_test_score)
print_metrics_table("Semantic RoBERTa Model", semantic_metrics)
```

```python
# =========================================================
# SECTION 13: FAST PERPLEXITY MODULE
# =========================================================

gpt2_name = "distilgpt2"   # faster than gpt2
gpt2_tokenizer = GPT2TokenizerFast.from_pretrained(gpt2_name)
gpt2_model = GPT2LMHeadModel.from_pretrained(gpt2_name).to(DEVICE)
gpt2_model.eval()

if gpt2_tokenizer.pad_token is None:
    gpt2_tokenizer.pad_token = gpt2_tokenizer.eos_token

def compute_text_nll(text, model, tokenizer, max_length=96):
    text = str(text)
    enc = tokenizer(
        text,
        truncation=True,
        max_length=max_length,
        return_tensors="pt"
    )
    input_ids = enc["input_ids"].to(DEVICE)
    attention_mask = enc["attention_mask"].to(DEVICE)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=input_ids)
        loss = outputs.loss.item()
    return loss

def compute_perplexity_scores(texts, model, tokenizer, max_length=96, desc="Perplexity"):
    nlls = []
    ppls = []
    for text in tqdm(texts, desc=desc):
        nll = compute_text_nll(text, model, tokenizer, max_length=max_length)
        ppl = math.exp(min(nll, 20))
        nlls.append(nll)
        ppls.append(ppl)
    return np.array(nlls), np.array(ppls)

train_nll, train_ppl = compute_perplexity_scores(train_df_fast["text_raw"], gpt2_model, gpt2_tokenizer, max_length=96, desc="Train PPL")
val_nll, val_ppl = compute_perplexity_scores(val_df_fast["text_raw"], gpt2_model, gpt2_tokenizer, max_length=96, desc="Val PPL")
test_nll, test_ppl = compute_perplexity_scores(test_df_fast["text_raw"], gpt2_model, gpt2_tokenizer, max_length=96, desc="Test PPL")

nll_mean = train_nll.mean()
nll_std = train_nll.std() if train_nll.std() > 1e-8 else 1.0

def nll_to_ai_score(nll_values, mean_val, std_val):
    z = (mean_val - nll_values) / std_val
    return 1 / (1 + np.exp(-z))

perplexity_train_score = nll_to_ai_score(train_nll, nll_mean, nll_std)
perplexity_val_score = nll_to_ai_score(val_nll, nll_mean, nll_std)
perplexity_test_score = nll_to_ai_score(test_nll, nll_mean, nll_std)

perplexity_metrics = classification_metrics(y_test_fast, perplexity_test_score)
print_metrics_table("Perplexity Module", perplexity_metrics)
```

```python
# =========================================================
# SECTION 14: FUSION MODEL
# =========================================================

fusion_train_X = np.column_stack([
    stylometry_train_score,
    semantic_train_score,
    perplexity_train_score
]).astype(np.float32)

fusion_val_X = np.column_stack([
    stylometry_val_score,
    semantic_val_score,
    perplexity_val_score
]).astype(np.float32)

fusion_test_X = np.column_stack([
    stylometry_test_score,
    semantic_test_score,
    perplexity_test_score
]).astype(np.float32)

class FusionMLP(nn.Module):
    def __init__(self, input_dim=3, hidden_dim=8, dropout=0.1):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, x):
        return self.net(x).squeeze(-1)

train_loader_fusion = DataLoader(NumpyDataset(fusion_train_X, y_train_fast), batch_size=64, shuffle=True)
val_loader_fusion = DataLoader(NumpyDataset(fusion_val_X, y_val_fast), batch_size=128, shuffle=False)

fusion_model = FusionMLP(input_dim=3, hidden_dim=8, dropout=0.1)
fusion_model = train_binary_torch_model(
    fusion_model,
    train_loader_fusion,
    val_loader_fusion,
    epochs=6,
    lr=1e-3,
    pos_weight=pos_weight
)

fusion_train_score = predict_binary_torch_model(fusion_model, fusion_train_X)
fusion_val_score = predict_binary_torch_model(fusion_model, fusion_val_X)
fusion_test_score = predict_binary_torch_model(fusion_model, fusion_test_X)

hybrid_metrics = classification_metrics(y_test_fast, fusion_test_score)
print_metrics_table("Hybrid Fusion Model", hybrid_metrics)
```
